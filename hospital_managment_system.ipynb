```python
class Person:
  def __init__(self,name,age,gender):
    self.name=name
    self.age=age
    self.gender=gender

class Patient(Person):
  def __init__(self,name,age,gender,patient_id,disease):
    super().__init__(name,age,gender)
    self.disease=disease
    self.patient_id=patient_id
    self.appointments=[]



  def get_details(self):
     return f"Name: {self.name} \npatient_id: {self.patient_id} \nAge: {self.age}\nGender: {self.gender}\nDisease: {self.disease}"




class Doctor(Person):
  def __init__(self, name, age, gender,speciality):
    super().__init__(name, age, gender)
    self.speciality=speciality



  def get_details(self):
    return f"Name: {self.name}  \nAge: {self.age}\nGender: {self.gender} \n speciality: {self.speciality}"



class Appointment:
  def __init__(self,doctor,patient,date):
    self.doctor=doctor
    self.patient=patient
    self.date=date

  def get_details(self):
        return f"Appointment on {self.date}:\nDoctor: {self.doctor.name}\nPatient: {self.patient.name}"

class Bill:
  def __init__(self,patient_id,serviece_fee):
    self.patient_id=patient_id
    self.serviece_fee=serviece_fee
  def generate_bill(self):
    return f"{self.patient_id} has to pay the {self.serviece_fee} service fee"



class Hospital:
  def __init__(self) :
    self.patients=[]
    self.doctors=[]
    self.appointments=[]

  def add_patient(self,patient):
    self.patients.append(patient)

  def discharge_patient(self, name):
        for p in self.patients:
            if p.name.lower() == name.lower():
                self.patients.remove(p)
                print(f" Patient '{name}' discharged successfully.")
                return
        print(f" Patient '{name}' not found.")

  def add_doctor(self, doctor):
        self.doctors.append(doctor)


  def schedule_appointment(self, patient_id, doctor_name, date):
      patient = next((p for p in self.patients if p.patient_id == patient_id), None)
      doctor = next((d for d in self.doctors if d.name.lower() == doctor_name.lower()), None)

      if patient and doctor:
          appt = Appointment(doctor, patient, date)
          self.appointments.append(appt)
          patient.appointments.append(appt)
          print(" Appointment scheduled.")
      else:
          print(" Doctor or Patient not found.")

  def view_all_patients(self):
      for p in self.patients:
        print(p.get_details(), "\n")

  def view_all_doctors(self):
      for d in self.doctors:
            print(d.get_details(), "\n")

  def view_all_appointments(self):
      for a in self.appointments:
          print(a.get_details(), "\n")







# Create hospital
h = Hospital()

# Add doctors
doc1 = Doctor("Dr. Smith", 40, "Male", "Cardiology")
doc2 = Doctor("Dr. Sara", 35, "Female", "Dermatology")
h.add_doctor(doc1)
h.add_doctor(doc2)

# Add patients
p1 = Patient("Asfand", 19, "Male", "P001", "Fever")
p2 = Patient("Zara", 22, "Female", "P002", "Cold")
h.add_patient(p1)
h.add_patient(p2)

# Schedule appointment
h.schedule_appointment("P001", "Dr. Smith", "2025-08-01")

# View all
h.view_all_patients()
h.view_all_doctors()
h.view_all_appointments()

# Generate bill
b = Bill("P001", 1500)
print(b.generate_bill())

# Discharge
h.discharge_patient("Asfand")


```
